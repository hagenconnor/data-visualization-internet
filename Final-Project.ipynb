#### Question: In the year 2012, which industry has the highest internet usage, and what type of connection do they use the most?


```python
#importing packages
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
#importing dataset
df = pd.read_csv("industry.csv")
#filtering columns and rows
df2 = df[df["REF_DATE"] == 2012]
df3 = df2[['REF_DATE', 'North American Industry Classification System (NAICS)', 'Size of enterprise', 'Type of Internet connection', 'VALUE']]
df3
```

```python
#renaming columns 
df4 = df3.rename(columns = {'REF_DATE':'Date'})
df5 = df4.rename(columns = {'North American Industry Classification System (NAICS)':'Class'})
df6 = df5.rename(columns = {'VALUE':'Value'})
df7 = df6.rename(columns = {'Type of Internet connection':'Type'})
df8 = df7.rename(columns = {'Size of enterprise':'Size'})
df8
```

Now that we have renamed our columns for easier access, and filtered out the rows for the year of 2012 as well as filter out unesscessary columns, we can compare specific industry to see what type of internet connection takes up their whole usage. 

```python
# taking a look at mobile wireless internet connection
df9 = df8[df8['Size'] == 'Small size enterprises']
df10 = df9.dropna()
df10
```

```python
#Sort highest to lowest.
df10 = df10.sort_values("Value",ascending= False)
df10
```

```python
SmallDSL = df10[df10['Type'] == 'Digital Subscriber Line']
SmallDSL = SmallDSL['Value'].mean()
SmallCable = df10[df10['Type']== 'Cable modem']
SmallCable = SmallCable['Value'].mean()
SmallWireless = df10[df10['Type']== 'Mobile wireless']
SmallWireless = SmallWireless['Value'].mean()

fig, axes = plt.subplots(figsize=(6,4))
plt.bar('Small DSL',SmallDSL)
plt.bar('Small Cable',SmallCable)
plt.bar('Small Wireless',SmallWireless)
```
